Question 1

Write a Python program to create a lambda function that adds 25 to a given number passed in as an argument.

sample input: 10

sample output: 35

```python
input1 = int(input("Enter Number (to add with 25) : "))
func = lambda X:X+25
print('The output after adding 25 is : ' , func(input1))
```
